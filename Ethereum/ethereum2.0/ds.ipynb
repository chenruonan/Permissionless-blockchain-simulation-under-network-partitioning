```python
# 节点node查询交易tx_hash是否上链 返回tx_receipt
def check_transaction_confirmation(node: Web3, contract: Contract, tx_hash):
    while True:
        try:
            tx_receipt = node.eth.get_transaction_receipt(tx_hash)
            print(
                "Transaction %s confirmed in block %d,节点%s中，计数器值为%d"
                % (
                    tx_hash,
                    tx_receipt["blockNumber"],
                    node.eth.accounts[0],
                    contract.functions.getCounter().call(),
                )
            )
            return tx_receipt  # 如果不为None，表示交易已经确认
        except Exception:
            print(
                "transaction %s not commited yet.节点%s中计数器值为%d"
                % (
                    tx_hash,
                    node.eth.accounts[0],
                    contract.functions.getCounter().call(),
                )
            )
            time.sleep(1)

# get the counter from node4
def getCounter(node1: Web3,  contract1: Contract):
    while 1:
        print(
            "%s 节点%s中，计数器值为%d"
            % (datetime.now(),node1.eth.accounts[0], contract1.functions.getCounter().call())
        )
        time.sleep(0.5)

thread_getCounter = threading.Thread(
    target=getCounter, args=( node1,  contract1)
)
thread_getCounter.start()

tx_hashlist=[]
for i in range(10):
    hash4 = contract1.functions.decrementByFour().transact().hex()
    tx_hashlist.append(hash4)
```
